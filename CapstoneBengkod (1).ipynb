2. Preprocessing Data

Point 1: Tangani missing values, error, duplikasi, dan outlier di setiap kolom.

```python
import pandas as pd
```

```python
# Membaca dataset
df = pd.read_csv('ObesityDataSet.csv')
```

```python
# Tampilkan 5 baris pertama
df.head()
```

```python
# Cek missing values
print("Jumlah missing values tiap kolom:\n", df.isnull().sum())
```

```python
# Drop baris dengan missing values
df = df.dropna()
```

```python
# Hapus data duplikat
df.drop_duplicates(inplace=True)
```

```python
# Tangani missing values (jika ada)
# Misal: untuk kolom numerik kita isi dengan median
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# Tangani outlier (menggunakan IQR)
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]
```

```python
# Fungsi untuk menghapus outlier dengan metode IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Pastikan kolom numerik bertipe float
numerical_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')

# Hapus outlier di setiap kolom numerik
for col in numerical_cols:
    df = remove_outliers_iqr(df, col)

print("Ukuran data setelah bersih:", df.shape)
```

Point 2: Ubah data kategori menjadi numerik

```python
# Tampilkan kolom bertipe object (kategori)
df.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder

# Buat instance encoder
le = LabelEncoder()
```

```python
# Terapkan ke semua kolom kategori
for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

Point 3: Tentukan apakah semua feature perlu digunakan

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Korelasi antar fitur")
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']

selector = SelectKBest(score_func=f_classif, k='all')  # atau ganti k sesuai jumlah fitur yang diinginkan
selector.fit(X, y)
```

```python
# Lihat skor setiap fitur
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': selector.scores_})
feature_scores.sort_values(by='Score', ascending=False)
```

Point 4: Atasi ketidakseimbangan kelas data

```python
df['NObeyesdad'].value_counts().plot(kind='bar', title='Distribusi Kelas')
plt.xlabel('Kelas')
plt.ylabel('Jumlah')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Sebelum SMOTE:", y.value_counts())
print("Setelah SMOTE:", pd.Series(y_resampled).value_counts())
```

Point 5: Lakukan normalisasi atau standarisasi data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Ubah kembali ke DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
df_final = pd.concat([X_scaled_df, pd.Series(y_resampled, name='NObeyesdad')], axis=1)
```

 Kesimpulan: Ini kita berhasil membersihkan data dengan menghilangkan missing values, duplikasi, dan outlier menggunakan metode IQR. Selanjutnya, data kategori diubah menjadi angka agar bisa diproses oleh model machine learning. Setelah itu, dilakukan seleksi fitur dengan melihat korelasi dan skor fitur untuk memastikan hanya data yang relevan yang digunakan.
